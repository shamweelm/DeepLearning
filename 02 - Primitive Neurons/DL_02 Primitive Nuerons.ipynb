# MP Nueron Class

```
class MPNueron:
    def __init__(self):
        self.b= None
    
    def model(self,x):  # defines how the model works for each row(x) i.e. mainly the mathematical formula
        return(sum(x)>=self.b)

    def predict(self,X): # Given a training/test set of data, this funcion is used to predict the output for dataset
        Y = [] 
        for x in X:
            result = self.model(x)
            Y.append(result) # store the result/ output in Y
        return np.array(Y)
    
    def fit(self,X,Y): # takes both X and Y dataset to compute best b and return the highest accuracy
        accuracy =  {}

        for b in range(X.shape[1] + 1):
            self.b = b
            Y_pred = self.predict(X)
            accuracy[b] = accuracy_score(Y, Y_pred)
        
        best_b = max(accuracy, key = accuracy.get) # see this!!!!!!
        self.b = best_b

        print("Best b value is = ", best_b)
        print("Highest Accuracy is = ", accuracy[best_b])



    
```

# Perceptron

```
type(X_train)
```

```
X_train = X_train.values
X_test = X_test.values
```

$y = 1, \mbox{if} \sum_i w_i x_i >= b$

$y = 0, \mbox{otherwise}$

```
class Perceptron:
    def __init__(self):
        self.w = None
        self.b = None

    def model(self, x):
        return 1 if(np.dot(self.w, x)>= self.b) else 0
    
    def predict(self, X):
        Y = []

        for x in X:
            result = self.model(x)
            Y.append(result)

        return np.array(Y)

    def fit(self, X, Y, epochs = 1, lr = 1): #lr and epochs are hyper-parameters and b,w are parameters

        self.w = np.ones(X.shape[1]) #size of W vector is equal to the number of columns in X
        self.b = 0

        accuracy = {}
        max_accuracy = 0

        wt_matrix = []

        for i in range(epochs):
            for x,y in zip(X,Y):
                y_pred = self.model(x)
                if y==1 and y_pred ==0:
                    self.w = self.w + lr * x
                    self.b = self.b + lr * 1
                elif y==0 and y_pred ==1:
                    self.w = self.w - lr * x
                    self.b = self.b - lr * 1
            accuracy[i] = accuracy_score(self.predict(X), Y)
            wt_matrix.append(self.w)
            if accuracy[i] > max_accuracy:
                max_accuracy = accuracy[i]
                chkw = self.w
                chkb = self.b
            
            
        
        self.w = chkw
        self.b = chkb
        print("Max Accuracy = ", max_accuracy)

        plt.plot(list(accuracy.values()))
        plt.ylim(0,1)
        plt.show() 
        return np.array(wt_matrix)  
```

```
perceptron = Perceptron()
```

```
wt_matrix = perceptron.fit(X_train, Y_train,100)
```

```
Y_pred_train = perceptron.predict(X_train)
print("Accuracy for train is = ", accuracy_score(Y_pred_train, Y_train))
```

```
Y_pred_test = perceptron.predict(X_test)
print("Accuracy for test is = ", accuracy_score(Y_pred_test, Y_test))
```
